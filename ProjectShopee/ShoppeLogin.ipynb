```python
links = []
pictures = []
name_products = []
prices = []
discounts = []
ratings = []
solds = []
```

```python
# li_tags = driver.find_elements(By.XPATH, "//ul[@class='row shopee-search-item-result__items']//li")
```

```python
# scroll()

# Lay thong tin san pham 
li_tags = driver.find_elements(By.XPATH, "//ul[@class='row shopee-search-item-result__items']//li")
for link in li_tags:
    try:
        link_element = link.find_element(By.XPATH, ".//a").get_attribute("href")
        links.append(link_element)
    except Exception as e:
        links.append("N/A")
        print(f"Loi {e}")
    

print(links)
print(len(links))
```

```python
# scroll()

# Tim anh san pham
li_tags = driver.find_elements(By.XPATH, "//ul[@class='row shopee-search-item-result__items']//li")
for picture in li_tags:
    try:
        picture_element = picture.find_element(By.XPATH, ".//div[@class='relative z-0 w-full pt-full']//img").get_attribute("src")
        pictures.append(picture_element)
    except Exception as e :
        pictures.append("N/A")
        print(f"Loi {e}")
        
print(pictures)
print(len(pictures))
```

```python
# scroll()

# Tim ten san pham
li_tags = driver.find_elements(By.XPATH, "//ul[@class='row shopee-search-item-result__items']//li")
for name in li_tags:
    try:
        name_element = name.find_element(By.XPATH, ".//div[@class='whitespace-normal line-clamp-2 break-words min-h-[2.5rem] text-sm']").text
        name_products.append(name_element)
    except Exception as e:
        name_products.append("N/A")
        print(f"Loi {e}")
        
print(name_products)
print(len(name_products))

```

```python
# scroll()

# Tim gia san pham
li_tags = driver.find_elements(By.XPATH, "//ul[@class='row shopee-search-item-result__items']//li")
for price in li_tags:
    try:
        # Bien gia gom don vi 'đ' và gia tien thanh danh sach
        price_elements = price.find_elements(By.XPATH, ".//div[@class='truncate flex items-baseline']//span")
        # Noi cac chuoi trong danh sach qua cac khoang trang
        price_text = "".join([element.text for element in price_elements])
        prices.append(price_text)
    except Exception as e:
        prices.append("N/A")
        print(f"Loi {e}")
        
print(prices)
print(len(prices))
```

```python
# scroll()

# Tim khuyen mai san pham
li_tags = driver.find_elements(By.XPATH, "//ul[@class='row shopee-search-item-result__items']//li")
for discount in li_tags:
    try:
        # discount_element = discount.find_element(By.CSS_SELECTOR, "div.text-shopee-primary.font-medium.bg-shopee-pink.py-0\\.5.px-1.text-sp10\\/3.h-4.rounded-\\[2px\\].shrink-0.mr-1 span").text
        discount_element = discount.find_element(By.CSS_SELECTOR, "div.text-shopee-primary.font-medium.bg-shopee-pink.py-0\\.5.px-1.text-sp10\\/3.h-4.rounded-\\[2px\\].shrink-0.mr-1").text
        discounts.append(discount_element)
    except Exception as e:
        discounts.append("N/A")
        print(f"Loi {e}")
        
print(discounts)
print(len(discounts))


```

```python
# scroll()

# Tim so luot danh gia
li_tags = driver.find_elements(By.XPATH, "//ul[@class='row shopee-search-item-result__items']//li")
for rating in li_tags:
    try:
        rating_element = rating.find_element(By.CSS_SELECTOR, "div.text-shopee-black87.text-xs\\/sp14.flex-none").text
        ratings.append(rating_element)
    except Exception as e:
        ratings.append("N/A")
        print(f"Loi {e}")
        
print(ratings)
print(len(ratings))
```

```python
# scroll()

# Tim so luot da ban
li_tags = driver.find_elements(By.XPATH, "//ul[@class='row shopee-search-item-result__items']//li")
for sold in li_tags:
    try:
        sold_element = sold.find_element(By.CSS_SELECTOR, "div.truncate.text-shopee-black87.text-xs.min-h-4").text
        solds.append(sold_element)
    except Exception as e:
        solds.append("N/A")
        print(f"Loi {e}")
        
print(solds)
print(len(solds))
```
